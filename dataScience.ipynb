```python
#lİNEER REGRESYON İÇİN HÜKÜMETTEN MEMNUN MUSUNUZ SORUSUNUN CEVABI OLAN HÜKÜMET SÜTUNUNA HAYİR YERİNE 0 EVET YERİNE 1 OLMAK ÜZERE
#YENİ BİR NEW_HÜKÜMET SÜTUNU EKLEDİM BURADA
hükümet_array = np.array(data["hükümet"])
for i in range(0,47):
    if hükümet_array[i] == "Hayir":
        hükümet_array[i] = 0
    else:
        hükümet_array[i] = 1
hükümet_array
data["new_hükümet"] = hükümet_array
data.head()
```

```python
#GÜNDEMİ TAKİP ETME DERECESİNE GÖRE, SİZCE HÜKÜMET PANDEMİYİ İYİ YÖNETİYOR MU SORUSUNUN İLİŞKİSİ 
gündem = data.iloc[:,4].values.reshape(-1,1)
hükümet = data.iloc[:,10].values.reshape(-1,1)
regression = LinearRegression()
regression.fit(gündem, hükümet)

plt.figure(figsize = (10,10))
plt.plot(gündem,regression.predict(gündem),color="blue")
plt.xlabel("GÜNDEMİ NE DERECE TAKİP ETMEKTESİNİZ?")
plt.ylabel("SİZCE HÜKÜMET PANDEMİYİ İYİ YÖNETİYOR MU? EVET= 1 / HAYIR= 0")
plt.show()
```

```python
#ÖRGÜN EĞİTİM Mİ? UZAKTAN EĞİTİM Mİ?
colors = ["gray", "red"]
plt.figure(figsize=(8,8))
plt.pie(x=data['eğitim'].value_counts(),labels=data['eğitim'].value_counts().index, autopct='%0.2f%%',colors = colors)
plt.title("EĞİTİMDE NİTELİK GÖZÜNÜNE ALINIRSA ÖRGÜN EĞİTİM Mİ? UZAKTAN EĞİTİM Mİ?")
plt.show()
```
