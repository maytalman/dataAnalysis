```python
#lNEER REGRESYON N HKMETTEN MEMNUN MUSUNUZ SORUSUNUN CEVABI OLAN HKMET STUNUNA HAYR YERNE 0 EVET YERNE 1 OLMAK ZERE
#YEN BR NEW_HKMET STUNU EKLEDM BURADA
hkmet_array = np.array(data["hkmet"])
for i in range(0,47):
    if hkmet_array[i] == "Hayir":
        hkmet_array[i] = 0
    else:
        hkmet_array[i] = 1
hkmet_array
data["new_hkmet"] = hkmet_array
data.head()
```

```python
#GNDEM TAKP ETME DERECESNE GRE, SZCE HKMET PANDEMY Y YNETYOR MU SORUSUNUN LKS 
gndem = data.iloc[:,4].values.reshape(-1,1)
hkmet = data.iloc[:,10].values.reshape(-1,1)
regression = LinearRegression()
regression.fit(gndem, hkmet)

plt.figure(figsize = (10,10))
plt.plot(gndem,regression.predict(gndem),color="blue")
plt.xlabel("GNDEM NE DERECE TAKP ETMEKTESNZ?")
plt.ylabel("SZCE HKMET PANDEMY Y YNETYOR MU? EVET= 1 / HAYIR= 0")
plt.show()
```
